```python
# Определить корни

from math import sqrt
from sympy import *
from sympy.plotting import plot
init_printing()
x = Symbol('x', domain=S.Reals, real=True)
f = 18*x**3+5*x**2 + 10*x-30
a = [r for r in solve(f, x) if r.is_real]
print('Корни уравнения: ', a)
plot(f, (x, -1, 1))
```

```python
# Найти интервалы, на которых функция возрастает
derivative_f = f.diff(x)
plot(derivative_f, x)
reply = [r for r in solve(derivative_f, x) if r.is_real]
if solve_univariate_inequality(derivative_f > 0, x):
    print('Функция возрастает на всей действительной оси х')
elif solve_univariate_inequality(derivative_f < 0, x):
    print('Функция убывает на всей действительной оси х')
else:
    print(f'Функция возрастает: solve_univariate_inequality(derivative_f > 0, x)\nФункция убывает: solve_univariate_inequality(derivative_f < 0, x)')
    
```

```python
print(f'f > 0 при {solve_univariate_inequality(f > 0, x)}')
```
